```python
import torch
from diffusers import DiffusionPipeline
```

# Creating a DiffusionPipeline with specified parameters

- Model: stabilityai/stable-diffusion-xl-base-1.0
- Torch data type: torch.float16
- Using SafeTensors for enhanced safety
- Variant: fp16 (16-bit floating-point precision)


```python
pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")
```

# Defining a prompt for image generation

```python
prompt = "An astronaut riding a golden horse to the moon"

images = pipe(prompt=prompt).images[0]
```
